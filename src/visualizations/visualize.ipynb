```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.csv')
data
```

```python
cat_cols=list(data.select_dtypes('object'))
num_cols=list(data.select_dtypes(['int','float']))
```

### Univariate Visualizations

```python
plt.figure(figsize=(10,15))
plot=1
for i in data.drop(cat_cols,axis=1):
    ax = plt.subplot(7,3,plot)
    sns.histplot(x=data[i],kde=True)
    plt.xlabel(i)
    plot+=1
plt.tight_layout()
plt.show()
```

### Bivariate Visualizations

```python
plt.figure(figsize=(20,21))
plot=1
for i in data.drop(cat_cols,axis=1):
    ax = plt.subplot(7,3,plot)
    sns.histplot(x=data[i],hue=data['PerformanceRating'],kde=True)
    plt.xlabel(i)
    plot+=1
plt.tight_layout()
plt.show()
```
