### 2. Bivariate Analysis
  - Analyzes two numerical/categorical columns or one categorical and one numerical column.
  - lineplot/scatterplot - Used to analyze two numerical columns
  - countplot(when used with hue) - Used to analyze two categorical columns
  - barplot,stripplot,swarmplot,violinplot,boxplot - Used to analyze one numerical and one categorical column.

```python
plt.figure(figsize=(10,5),facecolor="white")
sns.countplot(data=data,x="EmpDepartment",hue="PerformanceRating")
plt.title('Department Wise Performances')
plt.xlabel('EmpDept',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.tight_layout()
plt.show()
```
