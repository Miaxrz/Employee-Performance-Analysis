Why Decision Tree was modeled?

Decision Tree is one of the best algorithms because it is not sensitive to outliers or scaling, it can work so well without it. However, algorithm can overfit if not handled so hyperparameter tuning was initialized with the values that could reduce the overfit and yield good result with less error.

Why Random Forest?

Random Forest is one of the ensemble learning algorithms that builds multiple estimators by bootstrap sampling the data. It creates multiple decision trees and bootstrap aggregates the data for better performance. It aims to reduce the variance of the model, making it more stable and less prone to overfitting.

Why Gradient Boosting?

It is one of the boosting algorithm which builds multiple estimators sequentially on the data in such a way that the model learns on the errors of the previous model to emerge as a strong learner. It aims to reduce the bias of the model to avoid underfitting the data, improving its ability to capture complex patterns in the data.

Why AdaBoost?

AdaBoost trains a series of weak learners sequentially. In each iteration, the algorithm focuses on the examples that were misclassified by the previous learners. It adjusts the weights of the training examples, giving more weight to those that were misclassified and less weight to those that were correctly classified.
